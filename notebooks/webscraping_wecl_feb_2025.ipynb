```python
import pandas as pd

# Load all HTML tables from the given URL
url ='https://cricclubs.com/WECL/playerRankings.do?year=2024&league=all&tagId=null&clubId=6705'
html_tables = pd.read_html(url)

# Inspect all tables to find the right one
for i, table in enumerate(html_tables):
    print(f"Table {i}:")
    print(table.head())  # Print the first few rows to identify the structure
    print("\n")

# Assuming the correct table is at index 1 (adjust based on inspection)
df = html_tables[1]

# Print column names before adjustments
print("Original Columns:")
print(df.columns)

# Adjust headers if necessary
if 'Player' not in df.columns:
    # Set the first row as the header
    df.columns = df.iloc[0]
    # Drop the first row (now redundant as it's the header)
    df = df[1:]
    # Clean column names to remove extra spaces
    df.rename(columns=lambda x: x.strip(), inplace=True)

# Print adjusted column names
print("Adjusted Columns:")
print(df.columns)

# Access and print the 'Player' column (if it exists)
try:
    print(df['Player'].head())
except KeyError:
    print("The column 'Player' is still not found.")

# Save the DataFrame to a CSV file
df.to_csv('player_rankings_2024.csv', index=False)
print("CSV file saved as 'player_rankings_2024.csv'")
```
